# Scatter plots

```python
# scatter plot between Men and median salary
men=graduates_mjrs['Men']
median=graduates_mjrs['Median']
full_time=graduates_mjrs['Full_time']
women=graduates_mjrs['Women']
unemp=graduates_mjrs['Unemployed']
emp=graduates_mjrs['Employed']


fig= plt.figure(figsize=(14,10))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.scatter(men, median, c='red')
ax1.set_xlabel("MEN")
ax1.set_ylabel("Median Salary")
ax1.set_title("Median Salary for men ")


ax2.scatter(women, median, c='green')
ax2.set_xlabel("Women")
ax2.set_ylabel("Median Salary")
ax2.set_title("Median Salary for women ")


ax3.scatter(full_time, median, c='black')
ax3.set_xlabel("fulltime")
ax3.set_ylabel("Median Salary")
ax3.set_title("Median Salary for fulltime emps ")


ax4.scatter(emp, unemp, c='orange')
ax4.set_xlabel("Employed")
ax4.set_ylabel("Unemployed")
ax4.set_title("Employed and unemployed ")

plt.show()
```

# Histograms

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
c=['red','yellow','black','orange','red','yellow','black','orange']
fig = plt.figure(figsize=(20,30))
for r in range(0,8):
    ax = fig.add_subplot(4,2,r+1)
    ax = graduates_mjrs[cols[r]].hist(color=c[r],edgecolor='blue')
    ax.set_xlabel(cols[r])
    ax.set_title(cols[r])
    ax.grid(False)
```
